Q1. Convert the height and weight columns to numerical forms. [2 marks]

```python
# Before Conversion
print("Height and Weight columns Before Conversion:\n")
print("Height Values: \n",data['Height'].unique())
print("Weight Values: \n",data['Weight'].unique())
```

Following are the things noted:
1. Height Column contains values in terms of both cm and inches.
2. Weight contains values in kg and lbs

```python
# Function for the conversion
def convert_weight(weight_str):
    if 'kg' in weight_str:
        return int(weight_str.replace('kg', ''))
    elif 'lbs' in weight_str:
        return int(int(weight_str.replace('lbs', '')) / 2.20462)
    else:
        return None


def convert_height(height_str):
    if 'cm' in height_str:
        return int(height_str.replace('cm', ''))
    elif "'" in height_str:
        feet, inches = map(int, height_str.replace('"', '').split("'"))
        return int((feet * 12 + inches) * 2.54)
    else:
        return None
```

```python
# Applying the function to the respective columns
data['Height'] = data['Height'].apply(convert_height)

data['Weight'] = data['Weight'].apply(convert_weight)
```

```python
# After Conversion
print("Height and Weight columns After Conversion:\n")
print("Height Values: \n",data['Height'].unique())
print("Weight Values: \n",data['Weight'].unique())
```

Q3. 'Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value column is Million, so multiply the row values by 1,000,000, etc. [3 marks]

```python
# Check for values
data['Value'].unique()
```

```python
data['Wage'].unique()
```

```python
data['Release Clause'].unique()
```

Identified that 'K' and 'M' are the only two characters involved. So values with K is replaced with 'value*1000' and M is replaced with 'value*1000000'.

```python
# Creating a function for the conversion

def convert_value(num):
    if 'M' in num:
        return float(num.replace('M', '').replace('€', '')) * 1000000
    elif 'K' in num:
        return float(num.replace('K', '').replace('€', '')) * 1000
    else:
        return float(num.replace('€', ''))
```

```python
# Conversion of all columns: 'Value', 'Wage' and "Release Clause'.
data['Value'] = data['Value'].apply(convert_value)
data['Wage'] = data['Wage'].apply(convert_value)
data['Release Clause'] = data['Release Clause'].apply(convert_value)
```

```python
# Validate if conversion happened or not
print(data[['Name', 'Value', 'Wage', 'Release Clause']].head())
```

Q4. Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical. [3 marks]

```python
# To check for columns in dataframe that contain 'star' character.

def has_star(column):
    return any(str(value).find('★') != -1 for value in column)

star_columns = [col for col in data.columns if has_star(data[col])]

print("Columns with 'star' characters are:", star_columns)
```

```python
print("Display few data of star columns identified before replace * with ''.\n",data[star_columns].head())
```

```python
def convert_star_column(value):
    try:
        return int(value.replace('★', ''))
    except ValueError:
        return None

# Convert columns with 'star' characters to numerical values
data[star_columns] = data[star_columns].applymap(convert_star_column)
```

```python
# Display the updated star columns that doesn't has '*'.
print("Display few data of star columns identified After replace * with ''.\n",data[star_columns].head())
```
